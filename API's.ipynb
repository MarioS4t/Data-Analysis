```python
import requests
import pandas as pd
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
resp = requests.get(url)
resp.raise_for_status()
resp
```

```python
data = resp.json()
data[0]['title']
```

```python
issues = pd.DataFrame(data, columns=['numbers', 'title', 'labels', 'states'])
issues
```
